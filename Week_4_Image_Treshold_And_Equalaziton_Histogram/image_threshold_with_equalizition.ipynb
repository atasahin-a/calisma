```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread("./camera.jpg",0)

histr = cv2.calcHist([img_gray],[0],None,[256],[0,256])

plt.plot(histr)
plt.show()
```

```python
img_gray=cv2.imread("./camera.jpg",0)
img_eq_hist = cv2.equalizeHist(img_gray)

fig = plt.figure(figsize=(10,7))
rows = 1
colums = 2
fig.add_subplot(rows,colums,1)
plt.imshow(img_gray, cmap="gray")
plt.axis('off')
plt.title("Original image")

fig.add_subplot(rows,colums,2)
plt.imshow(img_eq_hist,cmap="gray")
plt.axis('off')
plt.title("Equalize Image")
```

```python
hist_org = cv2.calcHist([img_gray],[0],None,[256],[0,256])
hist_eq = cv2.calcHist([img_eq_hist],[0],None,[256],[0,256])

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(hist_org, color='b')
plt.title('Orijinal Grnt Histogram')

plt.subplot(1, 2, 2)
plt.plot(hist_eq, color='b')
plt.title('Histogram Eitleme Sonras Histogram')
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
img_adaptive_eq_hist = clahe.apply(img_gray)

fig = plt.figure(figsize=(10, 7))
rows = 1
colums = 3

fig.add_subplot(rows, colums, 1)
plt.imshow(img_gray, cmap="gray")
plt.axis('off')
plt.title('Original Image')

fig.add_subplot(rows, colums, 2)
plt.imshow(img_eq_hist, cmap="gray")
plt.axis('off')
plt.title('Eq Image')

fig.add_subplot(rows, colums, 3)
plt.imshow(img_adaptive_eq_hist, cmap="gray")
plt.axis('off')
plt.title('Adaptive Image')
```
